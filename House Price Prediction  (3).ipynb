# Determining the relationship between the house features and how the price varies

```python
numeric_df = df.select_dtypes(include='number')

corr_matrix = numeric_df.corr()

saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)

sorted_features = saleprice_corr.index.tolist()
sorted_corr_matrix = corr_matrix.loc[sorted_features, ['SalePrice']]

#Plot heatmap:
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, len(sorted_features) * 0.4))
sns.heatmap(sorted_corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, cbar=True)
plt.title("Features Sorted by Correlation with SalePrice", fontsize=14)
plt.tight_layout()
plt.show()
```

# Comparison of both models Linear Regression and Random Forest

```python
def evaluate_model_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return [mae, mse, rmse, r2]

# Get metrics for each model
lr_metrics = evaluate_model_metrics(y_test, y_pred_lr)
rf_metrics = evaluate_model_metrics(y_test, y_pred_rf)

# Create DataFrame for comparison
comparison_df = pd.DataFrame({
    'Linear Regression': lr_metrics,
    'Random Forest Regressor': rf_metrics
}, index=['MAE', 'MSE', 'RMSE', 'RÂ² Score'])

# Display the comparison table
print("\n--- Model Comparison Report ---")
print(comparison_df)
```

- Random Forest Regressor outperforms Linear Regression in all metrics.

- It achieves lower errors and a higher RÂ², meaning it explains more variance in house prices.

- Thus, Random Forest is recommended for production use.
